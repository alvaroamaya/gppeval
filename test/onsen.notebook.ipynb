# use onsen library

```python
import onsen as hp

gpp1 = hp.GeothermalPowerPlant()
tools = hp.Tools()
sim = hp.MonteCarloSimulation()

# in case change, restart python Kernel
# mode = 'Qt4Agg'
mode = 'notebook'
# mode = 'inline'
if mode is 'Qt4Agg':
    tools.set_backend_figure()
elif mode is 'notebook':
     %matplotlib nbagg   
else:
    pass
```

# 1) data from file

```python
gpp1 = tools.read_file_csv(fn_input='HishikariKagoshima_06Nov2017_001.csv')
#print gpp1
```

```python
sim.set_iterations(1000)
gpp1=sim.calc_energy_potential(gpp1)
```
